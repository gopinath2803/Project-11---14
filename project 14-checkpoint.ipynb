# Data Preprocessing:

```python
feature_cols = ['Serial No.', 'GRE Score', 'TOEFL Score', 'University Rating', 'SOP','LOR ', 'CGPA', 'Research']
x = ds[feature_cols]
x.head()
```

```python
# check the type and shape of X
print(type(x))
print(x.shape)
```

```python
y = ds['Chance of Admit ']
y.head()
```

```python
x.shape
```

```python
y.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=12)
```

```python
print('Observation - Your output should appear at this point as:')
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
lm = LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
lm.score(x_train,y_train)
```

# Observation:
The Percentage Score obtained by training the Linear Regression model is 83.18%

```python
pred=lm.predict(x_test)
print(pred)
```
